```python
import pandas as pd
data1 = pd.read_csv('C:/Users/Putra/Downloads/GOOG(1).csv')
```

```python
data1.sample(5)
```

```python
data1.drop(['Date'], axis = 1) #data1 tanpa kolom 'Date', axis = 1 menunjukan axis kolom (axis = 0 , menunjukan axis baris)
```

```python
data2 = data1
for i in data2.drop(['Date'], axis = 1).columns :
    data2[i] = data2[i] / data2[i].sum() # scalling data
```

```python
data2
```

```python
import matplotlib.pyplot as plt
data2[['Open','Close']].plot(x = 'Open', y='Close', style = 'o')
plt.title('Open Vs Close')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
